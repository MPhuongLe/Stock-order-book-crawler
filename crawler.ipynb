```python
PAGE_SIZE = 1500
SYMBOL_FILE = "symbol.txt"
```

```python
from vnstock import Vnstock
from datetime import datetime

# 1. đọc danh sách symbol
with open(SYMBOL_FILE, "r") as f:
    symbols = f.read().strip().split(",")

# thời gian và ngày hiện tại
now = datetime.now()
date_str = now.strftime("%d%m%Y")
time_str = now.strftime("%H%M")

for symbol in symbols:
    symbol = symbol.strip()

    try:
        # 2. khởi tạo đối tượng
        stock = Vnstock().stock(symbol=symbol, source='VCI')

        # 3. lấy dữ liệu intraday
        df = stock.quote.intraday(page_size = PAGE_SIZE)

        # 4. tạo tên file
        filename = f"{symbol}-{date_str}-{time_str}.csv"

        # 5. lưu file
        df.to_csv(filename, index=False)

        print(f"Saved {filename} ({len(df)} rows)")

    except Exception as e:
        print(f"Error with {symbol}: {e}")
```
